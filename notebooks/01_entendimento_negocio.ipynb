##### 1.1 Qual problema de negócio está sendo resolvido?

Problema abordado: **o NPS só é coletado após o encerramento da jornada de compra**

Solução proposta e para análise: Transformar dados operacionais (logística, atendimento, pedido, etc.) em **insights preditivos**, identificar gaps e correlação entre os dados para melhorar a tomada de decisão pela gestão.

Apesar do problema levantado no desafio ser o momento em que o NPS é coletado, como pode ser visto em alguns dos pontos de atenção que levantei, há muitas outras coisas que precisam levantadas e corrigidas antes de apenas de implementar soluções para a coleta do NPS em outros momentos da experiência do cliente.

##### 1.2 Por que o NPS é importante para um e-commerce?

O Net Promoter Score (NPS) é a métrica de lealdade mais usada no mundo e, para um e-commerce, tem relevância estratégica em três dimensões:

| Dimensão | Por que importa |
|----------|-----------------|
| **Lealdade do cliente** | Promotores (NPS 9-10) têm 2 a 5x mais chances de recompra e maior LTV (Lifetime Value) |
| **Crescimento orgânico** | Promotores indicam a marca para amigos e familiares, reduzindo o CAC (Custo de Aquisição de Clientes) |
| **Sinal operacional** | Quedas no NPS antecipam problemas de retenção, churn e reputação, antes que apareçam nas métricas financeiras |

Em e-commerce especificamente, onde a competição é intensa e a troca de plataforma é simples, o NPS diferencia empresas que retêm clientes das que apenas conquistam.

#### 2. Classificação do NPS - Contextualização

O NPS não é apenas uma nota, é uma **classificação de comportamento de cliente**:

| Nota | Categoria | Comportamento esperado |
|------|-----------|------------------------|
| 9 – 10 | **Promotor** | Extremamente satisfeitos que se tornam defensores fiéis da sua marca, alta fidelidade, maior LTV |
| 7 – 8 | **Neutro** | Satisfeito, mas vulnerável à concorrência |
| 0 – 6 | **Detrator** | Insatisfeito, risco de churn e de reviews negativos |

**Cálculo do NPS:**
```
NPS = % Promotores - % Detratores
```
O resultado varia de -100 (todos detratores) a +100 (todos promotores).

**Referências de mercado (e-commerce brasileiro):**
- NPS acima de 50: Excelente
- NPS entre 30 e 50: Bom
- NPS entre 0 e 30: Necessita atenção
- NPS negativo: Situação crítica

#### 3. Reflexão: Como o NPS Impacta o Negócio

##### 3.1 Impacto na Recompra

A relação entre NPS e recompra é direta e economicamente significativa:

- **Promotores (NPS 9-10):** Alta propensão a recompra, maior ticket médio ao longo do tempo e menor custo de retenção. Estudos de Reichheld (criador do NPS) mostram que aumentar a retenção em 5% pode elevar o lucro em até 25-95%.
- **Neutros (NPS 7-8):** Satisfeitos, mas vulneráveis a ofertas de concorrentes. Não geram recompra consistente nem indicações.
- **Detratores (NPS 0-6):** Alta probabilidade de churn, além de gerarem reclamações públicas que custam novas aquisições.

**Para o e-commerce em questão:** A variável `repeat_purchase_30d` presente nos dados confirma essa dinâmica -> clientes com NPS alto são mais propensos a recomprar dentro de 30 dias.

##### 3.2 Impacto no Boca a Boca

O boca a boca é o canal de aquisição com maior taxa de conversão e menor custo:

- **Promotores** recomendam ativamente, estudos apontam que cada promotor gera em média 2-3 novos clientes via indicação orgânica;
- **Detratores** fazem o oposto: cada reclamação em redes sociais ou marketplaces (ex.: Reclame Aqui, Google Reviews) pode inibir múltiplas compras;
- Em e-commerce, reviews e avaliações são parte crítica do funil de decisão, um NPS baixo se materializa em reviews negativos que afetam a conversão de novos visitantes.

**Equação de impacto:** Converter detratores em neutros e neutros em promotores reduz o CAC e aumenta o crescimento orgânico, reduzindo o investimento em mídia paga.

##### 3.3 Impacto no Market Share em E-commerce

O mercado de e-commerce brasileiro é altamente competitivo, com baixo custo de troca para o consumidor. Nesse contexto:

- Empresas com NPS alto crescem de forma mais sustentável, pois retêm clientes com menor investimento;
- Empresas com NPS baixo precisam investir mais em aquisição para compensar o churn;
- Grandes players como Magazine Luiza e Mercado Livre investem significativamente em experiência pós-venda justamente porque o NPS tem efeito direto no market share;
- Benchmarks indicam que e-commerces brasileiros líderes operam com NPS entre 40 e 65. Estar abaixo desse patamar sinaliza risco competitivo.

##### 3.4 Riscos do Negócio no E-commerce Brasileiro

O e-commerce nacional opera em um ambiente de riscos estruturais que impactam diretamente a satisfação do cliente e o NPS. Os principais são:

**Desigualdade Logística Regional**

A infraestrutura logística brasileira é concentrada nas regiões Sul e Sudeste. Clientes do Norte, Nordeste e Centro-Oeste enfrentam prazos de entrega maiores, mais tentativas de entrega e fretes mais caros - fatores diretamente ligados ao NPS. Isso cria uma desigualdade de experiência estrutural que não depende da operação central da empresa, mas do ecossistema logístico regional.

A variável `customer_region` e `delivery_delay_days` nos dados deste desafio refletem diretamente essa realidade.

**Alto Custo do Frete em Relação ao Ticket Médio**

O frete no Brasil representa um percentual elevado do valor dos pedidos, especialmente em compras de menor valor. A prática de "frete grátis" como alavanca competitiva pelos grandes players cria uma expectativa de mercado que deteriora o NPS quando o frete é cobrado, independentemente se o prazo seja cumprido.

A variável `freight_value` em relação a `order_value` permite investigar esse efeito nos dados disponíveis.

**Fraude, Chargeback e Cancelamentos Indevidos**

O Brasil está entre os países com maior índice de fraude no e-commerce da América Latina. Além do impacto financeiro direto, sistemas antifraude excessivamente restritivos podem bloquear pedidos legítimos, gerando cancelamentos indevidos que transformam clientes em detratores mesmo antes da entrega.

**Reputação Digital e Cultura de Reclamação (Reclame Aqui)**

O Reclame Aqui é uma plataforma de reclamações com impacto direto na decisão de compra no Brasil. Pesquisas mostram que parcela significativa dos consumidores brasileiros consulta a reputação da empresa na plataforma antes de finalizar uma compra. Uma nota baixa no RA1000 reduz a conversão e eleva o CAC.

O índice de reclamações no Reclame Aqui funciona como proxy externo do NPS, a variável `complaints_count` nos dados conecta diretamente esses dois mundos.

**Logística Reversa e Devoluções (Direito de Arrependimento — CDC)**

O Art. 49 do Código de Defesa do Consumidor garante ao cliente o direito de arrependimento em até 7 dias corridos para compras online, sem necessidade de justificativa. A logística reversa no Brasil é cara, lenta e pouco padronizada, gerando atrito considerável na experiência de devolução e impactando negativamente o NPS pós-venda.

**Sazonalidade Extrema e Estresse Operacional**

Datas como Black Friday, Dia das Mães, Dia dos Namorados e Natal concentram volumes massivos de pedidos em janelas muito curtas. O estresse operacional nessas datas degrada o SLA logístico e sobrecarrega o SAC.

**Concentração de Market Share e Pressão sobre Margem**

O e-commerce brasileiro é dominado por poucos grandes players (Mercado Livre, Shopee, Amazon, Americanas, Magazine Luiza), que impõem pressão de preço, prazo e experiência sobre os demais. Operadores menores que não conseguem competir em experiência tendem a compensar com descontos agressivos, criando expectativas desalinhadas que impactam negativamente o NPS pós-compra, como sugere a variável `discount_value` nos dados.

**Alta Tributação e Complexidade Fiscal**

O Brasil possui uma das cargas tributárias mais complexas do mundo para o varejo digital. A guerra fiscal do ICMS entre estados, a tributação sobre o frete e os encargos sobre marketplaces impactam diretamente a competitividade de preço e a capacidade de investimento em experiência do cliente.

**LGPD e Coleta de Dados de Satisfação**

A Lei Geral de Proteção de Dados (Lei nº 13.709/2018), em vigor desde 2021, impõe restrições à coleta, armazenamento e uso de dados pessoais, incluindo os utilizados nas pesquisas de NPS e nos modelos preditivos desenvolvidos neste projeto. A empresa precisa garantir que o consentimento para uso dos dados esteja adequado à LGPD, sob risco de sanções da ANPD (Autoridade Nacional de Proteção de Dados).

#### 4. NPS

##### 4.1 Qual variável representa a satisfação do cliente?

A variável alvo é **`nps_score`**, a nota do NPS atribuída pelo cliente, variando de **0 a 10**, coletada após a experiência de compra.

##### 4.2 Por que ela foi escolhida?

`nps_score` é a variável mais adequada por três razões:

1. **Padrão de mercado:** O NPS é a metodologia mais usada globalmente para medir lealdade e satisfação - comparável entre empresas, setores e períodos;
2. **Diretamente coletada do cliente:** Não é uma métrica calculada ou estimada - é a voz direta do consumidor;
3. **Preditora de comportamento futuro:** A nota NPS correlaciona-se com recompra, churn e indicações, tornando-a estrategicamente relevante.

As demais variáveis de satisfação presentes no dataset (`csat_internal_score`, `repeat_purchase_30d`) são complementares ou consequentes, não substitutas.

##### 4.3 Em que momento da jornada essa informação é coletada?

O `nps_score` é coletado **após o encerramento da jornada de compra**, tipicamente:

```
[Pedido] → [Pagamento] → [Processamento] → [Envio] → [Entrega] → [Coleta NPS]
```

**Não há informação de quanto tempo após o recebimento o NPS é coletado.**

É uma avaliação que enxerga o passado, não em tempo real - o que justifica a necessidade de um modelo preditivo que antecipe o NPS com base nos dados operacionais.

##### 4.4 Risco do uso da informação

**1. Viés de resposta:**
Clientes muito satisfeitos (promotores) e muito insatisfeitos (detratores) têm maior propensão a responder à pesquisa, ou seja, assim como eu, clientes neutros naturalmente não respondem, o que pode distorcer a distribuição do NPS coletado. Então o dataset pode super-representar extremos.

**2. Defasagem temporal:**
A pesquisa de NPS pode ser enviado dias ou semanas após a entrega e o cliente pode avaliar dias ou semanas após o recebimento, sendo assim, ele avalia a experiência com base na memória, que pode ser influenciada pelo estado emocional no momento da resposta, não apenas pela experiência objetiva.

**3. Score interno com `csat_internal_score`:**
Se o score interno de satisfação (`csat_internal_score`) utilizar o NPS como fonte de seu cálculo, utilizá-lo em uma análise de correlação com o NPS (mesmo que parcialmente) pode inflar artificialmente a performance do modelo quando essas duas variáveis estiverem disponíveis.

**4. Recompra de operações recentes com `repeat_purchase_30d`:**
A recompra em 30 dias pode ocorrer simultaneamente ou após o NPS, ou seja, esta variável pode não estar disponível no momento da predição para compras recentes, tornando os registros recentes inadequados para análise dessa métrica.

#### Conclusão

O problema está claro: o e-commerce precisa sair de uma postura **reativa** (medir NPS depois que a experiência aconteceu) para uma postura **preditiva e preventiva** (antecipar o NPS com base em dados operacionais e agir antes que o cliente se torne um detrator).

A variável `nps_score` é a target correta, desde que usada com os devidos cuidados para evitar data leakage e vieses de seleção. Os notebooks seguintes exploram os dados e constroem um modelo preditivo para apoiar essa transformação.
