##### 1.2 Por que o NPS é importante para um e-commerce?

O Net Promoter Score (NPS) é a métrica de lealdade mais usada no mundo e, para um e-commerce, tem relevância estratégica em três dimensões:

| Dimensão | Por que importa |
|----------|-----------------|
| **Lealdade do cliente** | Promotores (NPS 9-10) têm 2 a 5x mais chances de recompra e maior LTV (Lifetime Value) |
| **Crescimento orgânico** | Promotores indicam a marca para amigos e familiares, reduzindo o CAC (Custo de Aquisição de Clientes) |
| **Sinal operacional** | Quedas no NPS antecipam problemas de retenção, churn e reputação, antes que apareçam nas métricas financeiras |

Em e-commerce especificamente, onde a competição é intensa e a troca de plataforma é simples, o NPS diferencia empresas que retêm clientes das que apenas conquistam.

## 2. Reflexão: Como o NPS Impacta o Negócio

### 2.1 Impacto na Recompra

A relação entre NPS e recompra é direta e economicamente significativa:

- **Promotores (NPS 9-10):** Alta propensão a recompra, maior ticket médio ao longo do tempo e menor custo de retenção. Estudos de Reichheld (criador do NPS) mostram que aumentar a retenção em 5% pode elevar o lucro em até 25-95%.
- **Neutros (NPS 7-8):** Satisfeitos, mas vulneráveis a ofertas de concorrentes. Não geram recompra consistente nem indicações.
- **Detratores (NPS 0-6):** Alta probabilidade de churn, além de gerarem reclamações públicas que custam novas aquisições.

**Para o e-commerce em questão:** A variável `repeat_purchase_30d` presente nos dados confirma essa dinâmica — clientes com NPS alto são mais propensos a recomprar dentro de 30 dias.

---

### 2.2 Impacto no Boca a Boca

O boca a boca (word of mouth) é o canal de aquisição com maior taxa de conversão e menor custo:

- **Promotores** recomendam ativamente — estudos apontam que cada promotor gera em média 2-3 novos clientes via indicação orgânica;
- **Detratores** fazem o oposto: cada reclamação em redes sociais ou marketplaces (ex.: Reclame Aqui, Google Reviews) pode inibir múltiplas compras;
- Em e-commerce, reviews e avaliações são parte crítica do funil de decisão — um NPS baixo se materializa em reviews negativos que afetam a conversão de novos visitantes.

**Equação de impacto:** Converter detratores em neutros e neutros em promotores reduz o CAC e aumenta o crescimento orgânico, sem investimento adicional em mídia paga.

---

### 2.3 Impacto no Market Share em E-commerce

O mercado de e-commerce brasileiro é altamente competitivo, com baixo custo de troca para o consumidor. Nesse contexto:

- Empresas com NPS alto crescem de forma mais sustentável, pois retêm clientes com menor investimento;
- Empresas com NPS baixo precisam investir mais em aquisição para compensar o churn — espiral de custo crescente;
- Grandes players como Magazine Luiza e Mercado Livre investem significativamente em experiência pós-venda justamente porque o NPS tem efeito direto no market share;
- Benchmarks indicam que e-commerces brasileiros líderes operam com NPS entre 40 e 65. Estar abaixo desse patamar sinaliza risco competitivo.

## 3. Definição da Target

### 3.1 Qual variável representa a satisfação do cliente?

A variável alvo é **`nps_score`** — a nota de NPS atribuída pelo cliente, variando de **0 a 10**, coletada após a experiência de compra.

---

### 3.2 Por que ela foi escolhida?

`nps_score` é a variável mais adequada por quatro razões:

1. **Padrão de mercado:** O NPS é a metodologia mais usada globalmente para medir lealdade e satisfação — comparável entre empresas, setores e períodos;
2. **Diretamente coletada do cliente:** Não é uma métrica calculada ou estimada — é a voz direta do consumidor;
3. **Preditora de comportamento futuro:** A nota NPS correlaciona-se com recompra, churn e indicações, tornando-a estrategicamente relevante;
4. **Acionável:** Subir ou descer na escala NPS tem interpretação clara de negócio — cada ponto a mais representa um cliente mais próximo de ser promotor.

As demais variáveis de satisfação presentes no dataset (`csat_internal_score`, `repeat_purchase_30d`) são complementares ou consequentes, não substitutas.

---

### 3.3 Em que momento da jornada essa informação é coletada?

O `nps_score` é coletado **após o encerramento da jornada de compra**, tipicamente:

```
[Pedido] → [Pagamento] → [Processamento] → [Entrega] → [Uso do produto] → [NPS coletado]
```

Isso significa que a nota reflete a **experiência completa**: desde o checkout até a entrega e o eventual contato com o atendimento. É uma avaliação retrospectiva, não em tempo real — o que justifica a necessidade de um modelo preditivo que antecipe o NPS com base nos dados operacionais.

---

### 3.4 Existe algum risco de usar essa variável de forma inadequada?

Sim. Há quatro riscos principais:

**1. Viés de resposta (Non-response Bias):**
Clientes muito satisfeitos (promotores) e muito insatisfeitos (detratores) têm maior propensão a responder à pesquisa. Clientes neutros frequentemente não respondem, o que pode distorcer a distribuição do NPS coletado. O dataset pode super-representar extremos.

**2. Defasagem temporal:**
A pesquisa de NPS é enviada dias ou semanas após a entrega. O cliente avalia a experiência com base na memória, que pode ser influenciada pelo estado emocional no momento da resposta, não apenas pela experiência objetiva.

**3. Data Leakage com `csat_internal_score`:**
O score interno de satisfação (`csat_internal_score`) é fortemente correlacionado com o NPS. Se for calculado com base no NPS (mesmo que parcialmente), usá-lo como feature preditiva criaria vazamento de dados (leakage), inflando artificialmente a performance do modelo.

**4. Data Leakage com `repeat_purchase_30d`:**
A recompra em 30 dias pode ocorrer simultaneamente ou após o NPS. Em um modelo de produção — onde o NPS ainda não foi coletado — esta variável também pode não estar disponível no momento da predição, tornando-a inadequada como feature.

## 4. Classificação do NPS — Contextualização

O NPS não é apenas uma nota — é uma **classificação de comportamento de cliente**:

| Nota | Categoria | Comportamento esperado |
|------|-----------|------------------------|
| 9 – 10 | **Promotor** | Recomenda ativamente, alta fidelidade, maior LTV |
| 7 – 8 | **Neutro** | Satisfeito, mas vulnerável à concorrência |
| 0 – 6 | **Detrator** | Insatisfeito, risco de churn e de reviews negativos |

**Cálculo do NPS:**
```
NPS = % Promotores - % Detratores
```
O resultado varia de -100 (todos detratores) a +100 (todos promotores).

**Referências de mercado (e-commerce brasileiro):**
- NPS acima de 50: Excelente
- NPS entre 30 e 50: Bom
- NPS entre 0 e 30: Necessita atenção
- NPS negativo: Situação crítica
